# Exploring  dataset on the job outcomes of students who graduated from college between 2010 and 2012

In this project, I will be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github](https://github.com/fivethirtyeight/data/tree/master/college-majors) repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

* `Rank` - Rank by median earnings (the dataset is ordered by this column).
* `Major_code` - Major code.
* `Major` - Major description.
* `Major_category` - Category of major.
* `Total` - Total number of people with major.
* `Sample_size` - Sample size (unweighted) of full-time.
* `Men` - Male graduates.
* `Women` - Female graduates.
* `ShareWomen` - Women as share of total.
* `Employed` - Number employed.
* `Median` - Median salary of full-time, year-round workers.
* `Low_wage_jobs` - Number in low-wage service jobs.
* `Full_time` - Number employed 35 hours or more.
* `Part_time` - Number employed less than 35 hours.

```python
recent_grads.plot(x='Sample_size', 
y='Median',kind='scatter', title='Median vs. Sample_size')
```

From the above scatter plot, shows a weak correlation between median salary of full-time, year-round workers and the sample size of full time. There two outliers at over 4000 sample at approx 40k and at 100 sample size at over 100k.

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',
kind='scatter',title='Unemployment_rate vs. Sample_size')
```

From the above scatter plot, also shows a weak correlation between Unemployment rate and the sample size of full time.

```python
recent_grads.plot(x='Full_time', 
y='Median',kind='scatter', title='Full_time vs. Median')
```

From the above scatter plot, also shows a weak correlation between the number employed 35 hours or more and median salary of full-time, year-round workers

```python
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',
kind='scatter', title='ShareWomen vs. Unemployment_rate')
```

From the above scatter plot, shows no correlation between the women as share of total and the number employed 35 hours or more.

```python
recent_grads.plot(x='Men',y='Median',
kind='scatter', title='Men vs. Median')
```

```python
recent_grads.plot(x='Women',y='Median',
kind='scatter', title='Women vs. Median')
```
